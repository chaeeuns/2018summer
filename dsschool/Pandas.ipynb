# 컬럼 추가 & 수정

```python
#컬럼 값이 모두 동일 할 때
train["Category"] = "Titanic"
train.head()
```

```python
train["Id"] = range(891)
train.head()
```

```python
train["FamiliySize"] = train["SibSp"] + train["Parch"] + 1
train[["SibSp","Parch","FamiliySize"]].head()
```

```python
#조건마다 다르게 주기 
train["France"] = train["Embarked"] == "C"
train["UK"] = train["Embarked"] != "C"

train[["Embarked","France","UK"]].head()
```

```python
train.loc[train["Embarked"]=="C","Nationality"] = "France"
train.loc[train["Embarked"]!="C","Nationality"] = "UK"

train[["Embarked","Nationality"]].head()
```

```python
mean_age = train["Age"].mean()
train.loc[train["Age"].isnull(), "Age"] = mean_age
train.head(10)
```

```python
train.to_csv("train_pandas.csv")
```
