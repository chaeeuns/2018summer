# Name

```python
# 성,호칭.이름 패턴 확인
train["Name"].head()
```

```python
# 함수 get_title : 이름을 받아서 타이틀(호칭)을 반환함
def get_title(name):
    # name.split(",")[1] == Mrs.JohnBradly
    # ,,,.split(".")[0] == Mrs
    return name.split(",")[1].split(".")[0]
```

```python
train["title"] = train["Name"].apply(get_title)
train[["Name","title"]].head(10)
```

```python
sns.countplot(data = train, x="title")
# 주요 호칭이 Mr Mrs Mis Master 임을 확인
```

```python
sns.countplot(data=train, x="title", hue="Survived")
```

# Processing

```python
# Processing Sex Fare Embarked
```

# Encode Sex

```python
# Sex column 숫자로 바꾸기
train.loc[train["Sex"] == "male", "Sex_encode"] = 0
train.loc[train["Sex"] == "female", "Sex_encode"] = 1
print(train.shape)
train.head()
```

```python
test.loc[test["Sex"] == "male", "Sex_encode"] = 0
test.loc[test["Sex"] == "female", "Sex_encode"] = 1
print(test.shape)
test.head()
```

# Fill in missing fare

```python
train[train["Fare"].isnull()]
```

```python
test[test["Fare"].isnull()]
```

```python
train["Fare_fillin"] = train["Fare"]
test["Fare_fillin"] = test["Fare"]
test.loc[test["Fare_fillin"].isnull(),"Fare_fillin"] = 0
```

```python
test.loc[test["Fare"].isnull(),["Fare","Fare_fillin"]]
```

# Encode Embarked

```python
# One Hot Encoding
```

```python
train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"
train[["Embarked","Embarked_C","Embarked_S","Embarked_Q"]].head(10)
```

```python
test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"
test[["Embarked","Embarked_C","Embarked_S","Embarked_Q"]].head(10)
```

# Age

```python
# Feature Engineering 
# 숫자 데이터를 넣을 때 구역 나누기를 Decision Tree에게 맡기지 말고 내가 나누면 더 정확할 수 있다.
# 앞에서 데이터 시각화를 통해 얻은 정보 : 15세 이하 승객은 생존률 높다
```

```python
train["Child"] = train["Age"]<15
train[["Age","Child"]].head(10)
```

```python
test["Child"] = test["Age"]<15
test[["Age","Child"]].head(10)
```

# FamiliySize

```python
# Age와 마찬가지로 내가 구역 나눠주기
```

```python
test["familiysize"] = test["SibSp"] + test["Parch"] +1
test[["familiysize","SibSp","Parch"]].head()
# train은 데이터 시각화 할 때 만들어줬었음
```

```python
train["Single"] = train["familiysize"] == 1
train["Nuclear"] = (train["familiysize"] >1) & (train["familiysize"]<5)
train["Big"] = train["familiysize"] >= 5

train[["familiysize","Single","Nuclear","Big"]].head(10)
```

```python
test["Single"] = test["familiysize"] == 1
test["Nuclear"] = (test["familiysize"] >1) & (test["familiysize"]<5)
test["Big"] = test["familiysize"] >= 5

test[["familiysize","Single","Nuclear","Big"]].head(10)
```

# Name

```python
# 앞에서 Master가 남자지만 예외적으로 생존률이 높다는 사실을 확인했다.
# Master column을 만든다
```

```python
train["Master"] = train["Name"].str.contains("Master")
train[["Name","Master"]].head(10)
```

```python
test["Master"] = test["Name"].str.contains("Master")
test[["Name","Master"]].head()
```

# Train

```python
# Feature : Pclass, Sex_encode, Fare_fillin, Embarked(_CSQ), Age(Child), familiysize(Type)
# Label : Survived
```

```python
feature_names = ["Pclass","Sex_encode","Fare_fillin","Embarked_C","Embarked_S","Embarked_Q","Child","Single","Nuclear","Big","Master"]
feature_names
```

```python
label_name = "Survived"
label_name
```

```python
X_train = train[feature_names]
X_train.head()
```

```python
X_test = test[feature_names]
X_test.head()
```

```python
y_train = train[label_name]
y_train.head()
```

# Use Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth = 8, random_state = 0)
```

```python
model.fit(X_train,y_train)
```

# Visualize

```python
import graphviz
from sklearn.tree import export_graphviz
dot_tree = export_graphviz(model, feature_names = feature_names,
                           class_names = ["Perish","Survived"],
                           out_file=None)
graphviz.Source(dot_tree)
```

# Predict

```python
predictions = model.predict(X_test)
print(predictions.shape)
predictions[0:10]
```

# Submit

```python
submission = pd.read_csv("gender_submission.csv", index_col = "PassengerId")
print(submission.shape)
submission.head()
```

```python
submission["Survived"] = predictions
print(submission.shape)
submission.head()
```

```python
submission.to_csv("decision-tree_0.81818.csv")
```
